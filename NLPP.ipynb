```python
from nltk.corpus import stopwords
```

```python
stwords = set(stopwords.words('english'))
```

```python
print(stwords)
```

```python
def remove_stopwords(text,stopwords):
    usefl = [ w for w in text if w not in stopwords ]
    return usefl
```

```python
text = "I don't wanna do any thing today".split()
remove_stopwords(text,stwords)
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize,RegexpTokenizer
```

```python
regex = RegexpTokenizer('[a-zA-Z]+')
regex.tokenize("I don't wanna do any thing today")
```

```python
def tokenizer_custom(sentence):
    sent = regex.tokenize(sentence.lower())
    sent = remove_stopwords(sent,stwords)
    sent = [ps.stem(w) for w in sent]
    return sent
```

```python
tokenizer_custom("I don't Wanna do any thing annoying today")
```

```python
cv2 = CountVectorizer(tokenizer=tokenizer_custom)
vecdat = cv2.fit_transform(data_).toarray()
```

```python
cv2.inverse_transform(vecdat)
```
